```python
class CustomDataset(Dataset):
    def __init__(self, datapath, typeofdata=None, transform=None):
        self.datapath = datapath
        self.transform = transform
        self.typeofdata = typeofdata
        path = os.path.join(self.datapath,self.typeofdata)
        num_dirs = len(os.listdir(path))
        self.pairs = [os.path.join(path, "dir_" + str(d)) for d in range(num_dirs)]

    def __len__(self):
        return len(self.pairs)
        
    def __getitem__(self, idx):
        pair_dir = self.pairs[idx]
        img1_path = os.path.join(pair_dir, 'img1.jpg')
        img2_path = os.path.join(pair_dir, 'img2.jpg')
        image1 = read_image(img1_path)
        image2 = read_image(img2_path)

        if self.transform:
            image1 = self.transform(image1)
            image2 = self.transform(image2)
        if(self.typeofdata == "Positive"):
            label = 1
        elif(self.typeofdata == "Negative"):
            label = 0
        return image1, image2, label
```

```python
datapath = r"C:\Users\Naama Stahl\OneDrive - post.bgu.ac.il\שולחן העבודה\Physics\Year 5\Deep learning\HW #2\Data"
transform = transforms.Compose([
            transforms.Resize((125, 125)),
        ])
```

```python
dataset = CustomDataset(datapath,"Negative", transform=transform)
```
